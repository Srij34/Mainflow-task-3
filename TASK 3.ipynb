```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:\\Users\\Srijani\\Desktop\\Internship TASK\\householdtask3.csv')
```

```python
data.head(10)
```

```python
#scatter plot with year against own
plt.scatter(data['year'], data['own'])

#Adding title to the plot
plt.title("Scatter plot")

#Setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

#Showing the result
plt.show()
```

```python
#Line chart with year against own
plt.plot(data['year'])
plt.plot(data['own'])

#Adding title to the plot
plt.title("Line Chart")

#Setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

#Showing the result
plt.show()
```

```python
#Bar chart or Bar plot
plt.bar(data['year'], data['own'])

#Adding title to the plot
plt.title("Bar Chart")

#Setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

#Showing the result
plt.show()
```

```python
#Histogram
plt.hist(data['income'])

plt.title("Histogram")

plt.show()
```
